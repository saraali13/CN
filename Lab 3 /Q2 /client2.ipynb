```python
import socket
import threading

s = socket.socket()
s.connect(('localhost', 9997))

print("Connected to chat server")

def receive():
    while True:
        try:
            msg = s.recv(1024).decode()

            if not msg or msg.lower() == "exit":
                print("Server disconnected")
                s.close()
                break

            print("MSG: ", msg)
        except:
            break

threading.Thread(target=receive).start() #can receive multiple msgs 

while True:
    msg = input("Client: ")
    s.send(msg.encode())
    
    if msg.lower() == "exit":
        print("Disconnected by client")
        s.close()
        break
```
