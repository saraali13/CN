```python
import socket
import threading

s = socket.socket()
print("Socket created")

s.bind(('localhost', 9997))
s.listen(5)

print("Waiting for connections...")

clients = [] #stores all connected clients

#function used per thread
def handle_client(c, addr):
    print("Connected with", addr)
    clients.append(c)

    while True:
        try:
            msg = c.recv(1024).decode()

            if not msg or msg.lower() == "exit":
                print("Client disconnected:", addr)
                clients.remove(c)
                c.close()
                break

            print(f"{addr} says:", msg)

            for client in clients: #send message to all other clients except c 
                if client != c:
                    client.send(f"MSG from {addr}:{msg}".encode())

        except: #close the socket if connection fails or msg passing is intrupted 
            clients.remove(c)
            c.close()
            break

while True: #accept clients
    c, addr = s.accept()
    threading.Thread(target=handle_client, args=(c, addr)).start()
```
