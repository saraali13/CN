```python
import socket

s = socket.socket()
print("Socket created")

s.bind(('localhost', 9998))
s.listen(5)

print("Waiting for connection...")

while True:
    c, addr = s.accept()
    print("Connected with", addr)

    while True:
        # Receive message from client
        data_from_client = c.recv(1024).decode()

        #if clients send a message    
        if data_from_client.startswith("MSG:"):
            msg = data_from_client[4:] #slice the msg from index 4 to the end

            if not msg or msg.lower() == "exit":
                print("Client disconnected")
                c.close()
                break
            
            print("Client: ", msg)

            # Server reply
            reply = input("Server: ")
    
            c.send(reply.encode())

            if reply.lower() == "exit":
                print("Connection closed by server")
                c.close()
                break
                
        #if clients send a file    
        elif data_from_client.startswith("FILE:"):
            file = data_from_client[5:] #slice the msg from index 4 to the end
            
            print("Client send a File named: ", file)
            file_data = c.recv(4096)

            with open(file,"wb") as filep:
                filep.write(file_data)
                
            # Server reply
            reply = input("Server: ")
    
            c.send(reply.encode())

            if reply.lower() == "exit":
                print("Connection closed by server")
                c.close()
                break
```
