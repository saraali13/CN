```python
import socket

s = socket.socket()
s.connect(('localhost', 9998))

print("Connected to server")

while True:
    # Client sends message or file
    type_ = input("Send message (m/M) or file (f/F)?: ")
    s.send(type_.encode())

    #client sends a message 
    if type_ == "m" or type_ == "M":
        msg = input("Client: ")
        s.send(f"MSG: {msg}".encode())

        if msg.lower() == "exit":
            print("Disconnected from client")
            s.close()
            break
        
        # Receive reply from server
        reply = s.recv(1024).decode()

        if not reply or reply.lower() == "exit":
            print("Server ended the chat")
            s.close()
            break

        print("Server:", reply)

    #client sends a message 
    elif type_ == "f" or type_ == "F":
        filename = input("Enter filename: ")

        s.send(f"FILE: {filename}".encode())
        
        with open(filename,"rb") as filep:
            s.send(filep.read())
            
        print("file sent")
        
        # Receive reply from server
        reply = s.recv(1024).decode()
        print("Server:", reply)
        
        if reply.lower() == "exit":
            print("Disconnected from server")
            s.close()
            break
        


```
